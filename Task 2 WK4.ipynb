```python
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
class LoginPageTest:
    def __init__(self):
        # Configure Chrome to run in headless mode for environments without a display
        options = webdriver.ChromeOptions()
        options.add_argument('--headless')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        self.driver = webdriver.Chrome(options=options)
        self.base_url = "https://example.com/login"  # Replace with actual login page
        self.wait = WebDriverWait(self.driver, 10)

    def test_valid_login(self):
        """Test login with valid credentials"""
        try:
            self.driver.get(self.base_url)

            # AI-generated selectors for better element identification
            username_field = self.wait.until(EC.presence_of_element_located(
                (By.CSS_SELECTOR, "[data-testid='username']")))
            password_field = self.driver.find_element(By.CSS_SELECTOR, "[data-testid='password']")
            login_button = self.driver.find_element(By.CSS_SELECTOR, "[data-testid='login-btn']")

            username_field.send_keys("valid_user@example.com")
            password_field.send_keys("correct_password")
            login_button.click()

            # Verify successful login
            dashboard = self.wait.until(EC.presence_of_element_located(
                (By.CSS_SELECTOR, ".dashboard")))
            print("✓ Valid login test: PASSED")
            return True

        except Exception as e:
            print(f"✗ Valid login test: FAILED - {e}")
            return False

    def test_invalid_login(self):
        """Test login with invalid credentials"""
        try:
            self.driver.get(self.base_url)

            username_field = self.wait.until(EC.presence_of_element_located(
                (By.CSS_SELECTOR, "[data-testid='username']")))
            password_field = self.driver.find_element(By.CSS_SELECTOR, "[data-testid='password']")
            login_button = self.driver.find_element(By.CSS_SELECTOR, "[data-testid='login-btn']")

            username_field.send_keys("invalid_user@example.com")
            password_field.send_keys("wrong_password")
            login_button.click()

            # Verify error message
            error_message = self.wait.until(EC.presence_of_element_located(
                (By.CSS_SELECTOR, "[data-testid='error-message']")))
            print("✓ Invalid login test: PASSED")
            return True

        except Exception as e:
            print(f"✗ Invalid login test: FAILED - {e}")
            return False

    def run_tests(self):
        """Execute all test cases"""
        print("Running Login Page Tests...")
        valid_result = self.test_valid_login()
        time.sleep(2)
        invalid_result = self.test_invalid_login()

        success_rate = (valid_result + invalid_result) / 2 * 100
        print(f"\nTest Results:")
        print(f"Success Rate: {success_rate}%")
        print(f"Tests Passed: {valid_result + invalid_result}/2")

        self.driver.quit()
        return success_rate

# Execute tests
if __name__ == "__main__":
    test_suite = LoginPageTest()
    test_suite.run_tests()
```
